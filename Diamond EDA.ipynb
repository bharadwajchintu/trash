## Bar Plot

```python
plt.figure(figsize=(10,8))

sns.barplot(data = diamond_df,x = 'color', y = 'price',ci = None);
```

## Box Plot

```python
plt.figure(figsize=(12,8))

sns.boxplot(data = diamond_df, x = 'color', y = 'price');
```

### Distrubution Plot

```python
plt.figure(figsize=(10,8), dpi=700);

sns.displot(data = diamond_df,x = 'price',height=10,aspect=15/5);

plt.ylabel('price', fontdict={'fontsize':40,'fontweight':'bold', 'color':'purple'})
plt.xlabel('weight',fontdict={'fontsize':40,'fontweight':'bold', 'color':'purple'});
```

# KDE Plot

```python
plt.figure(figsize=(15,5))
sns.kdeplot(x = diamond_df.price);
```

```python
plt.figure(figsize=(8,10));

sns.displot(data =diamond_df, x = 'price',height = 8,aspect = 15/5,kde = True);

plt.ylabel('count', fontdict={'fontsize':40,'fontweight':'bold', 'color':'purple'})
plt.xlabel('price',fontdict={'fontsize':40,'fontweight':'bold', 'color':'purple'});
```
